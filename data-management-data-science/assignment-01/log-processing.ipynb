### Valid Records? 
Write a function that inspects a log line and returns **`True`** if the line appears to contain both a **valid timestamp** and at least one **action token**, and **`False`** otherwise. A timestamp should be considered valid if the line includes a date in either of the formats **`YYYY-MM-DD`** (e.g., `2026-01-03`) or **`YYYY/MM/DD`** (e.g., `2026/01/03`). If neither date format is present, the line should automatically return `False`. If a valid date is detected, the function should then verify that the line contains at least one additional non-date, non-whitespace token that can reasonably represent an action (such as a verb, command, status, or event label). This ensures the line is not just a timestamp but represents a meaningful log event rather than a standalone date string.


```python
# complete the function below
import random
def is_probably_valid(line):
    """
    Return True if the line looks like it contains a timestamp and at least an action. Assume that the time stamp is valid
    if it has at least a date in the foramt 2026-01-03 or 2026/01/03
    Otherwise False.
    """
    ## COMPLETE THE CODE HERE
    if line is None:
        return False
        
    line = line.strip()
    
    if line == "BAD LINE WITHOUT FIELDS" or line == "":
        return False

    valid_date = False
    valid_action = False
    
    tokens = line.split()
    
    for token in tokens:
        if len(token) >= 10:
            prefix = token[:10]

            if prefix[:4].isdigit() and prefix[4] in "-/":
                seperator = prefix[4]
                if prefix[5:7].isdigit() and prefix[7] == seperator and prefix[8:10].isdigit():
                    valid_date = True
                    continue

        
        clean = token.strip("|").lower()

        if clean.startswith("action"):
            valid_action = True
            break

        if clean in ("login", "view-page", "view_page", "download", "submit", "unknown"):
            valid_action = True
            break

    return valid_date and valid_action
    
### write some test code to check the function. You should select 5 random lines from the file read and determine if the line is valid
sample = random.sample(split_lines, 5) if len(split_lines) >= 5 else split_lines

print("5 random line checks:")
for ln in sample:
    print(is_probably_valid(ln), "|", ln)

### find the percentage of valid lines in the file

total_lines = 0
valid_lines = 0
for line in split_lines:
    total_lines += 1
    if is_probably_valid(line) == True:
        valid_lines += 1
percentage = (valid_lines / total_lines) * 100

print("Total percentage of valid lines: ", percentage,"%")
```

### Normalize Time Stamps
This timestamp normalization function standardizes multiple common date–time formats into a single consistent representation for reliable data processing and analysis. It converts timestamps such as **`2026-01-21 09:12:03`**, **`2026/01/21 09:13`**, and **`2026-01-21T09:15:40Z`** into the unified format **`YYYY-MM-DD HH:MM:SS`** (e.g., `2026-01-21 09:15:40`). By enforcing a single canonical structure, the function ensures compatibility across datasets, simplifies sorting, filtering, indexing, and time-series analysis, and prevents format-based parsing errors. When seconds are missing or invalid, they are automatically normalized to `"00"`, preserving temporal consistency while maintaining data integrity—an essential step for building clean, comparable, and machine-readable log data pipelines in data science and data engineering workflows.


```python
## time stamps normalization
def normalize_timestamp(ts):
    """
    Convert:
    - '2026-01-21 09:12:03'
    - '2026/01/21 09:13'
    - '2026-01-21T09:15:40Z'
    into '2026-01-21 09:15:40'
    Missing/invalid seconds become '00'
    """
    ts = ts.strip()
    if ts == "BAD LINE WITHOUT FIELDS" or ts == "":
        return None
        
    ts = ts.replace("T", " ")
    if ts.endswith("Z"):
        ts = ts[:-1]

    tokens = ts.split(" ")

    if len(tokens) == 0:
        return None

    date = tokens[0].strip()
    time = ""

    if len(tokens) >= 2:
        time = tokens[1].strip()
        
    if len(date) != 10:
        return None
    if not date[:4].isdigit():
        return None
    if date[4] != "-" and date[4] != "/":
        return None
    seperator = date[4]
    if not date[5:7].isdigit():
        return None
    if date[7] != seperator:
        return None
    if not date[8:10].isdigit():
        return None

    date = date.replace("/", "-")

    hours = "00"
    minutes = "00"
    seconds = "00"

    if time != "":
        t = time.split(":")
        if len(t) >= 2 and len(t[0]) == 2 and len(t[1]) == 2 and t[0].isdigit() and t[1].isdigit():
            hours = t[0]
            minutes = t[1]
            if len(t) >= 3 and len(t[2]) == 2 and t[2].isdigit():
                seconds = t[2] 
            else:
                seconds = "00"
            
    return date + " " + hours + ":" + minutes + ":" + seconds   
    

### write some test code to check the function
print(normalize_timestamp("2026/01/21 04:28"))
print(normalize_timestamp("2026-01-21T08:32Z"))
```

The function normalizes keys and values (e.g., case, spacing, separators), resolves semantic equivalences across different field names, and returns the results in a structured form such as a dictionary or a list of key–value pairs.
```python
## Key-value parsing 

def parse_fields(line):
    """
    Extract fields from messy formats like:
    'user=adg133' OR 'user : js3040'
    'action= view-page' OR 'action=VIEW-PAGE'
    'page=/x' OR 'file= y' OR 'assignment=lab1'
    Return a dict-like structure OR list-of-pairs (your choice).
    """
    fields = {}
    
    parts = line.split("|")

    for part in parts:
        part = part.strip()

        if part == "":
            continue

        words = part.split()

        i = 0
        
        while i < len(words):
            token = words[i]
        

            if "=" in token:
                key, value = token.split("=", 1)
                key = key.strip().lower()
                value = value.strip()

                if value == "" and i + 1 < len(words):
                    value = words[i + 1].strip()
                    i += 1

                    if key == "action" and value.lower() == "view" and i + 1 < len(words):
                        next_word = words[i + 1].strip()
                        if next_word.lower() == "page":
                            value = value + " " + next_word
                            i += 1

                if key != "" and value != "":
                    fields[key] = value
                    
            elif i + 2 < len(words) and words[i + 1] == ":":
                key = token.strip().lower()
                value = words[i + 2].strip()

                if key == "action" and value.lower() == "view" and i + 3 < len(words):
                    next_word = words[i + 3].strip()
                    if next_word.lower() == "page":
                        value = value + " " + next_word
                        i += 1
                
                if key != "" and value != "":
                    fields[key] = value
                i += 2
            i += 1

    return fields
### write some test code to check the function

tests = [
    "2026-01-21 12:33:34 | user=aa2573 | action=VIEW-PAGE | page=/courses/cs439",
    "2026-01-21T13:37:00Z | user : mg1773 | action= view-page | page=/profile",
    "BAD LINE WITHOUT FIELDS",
    "2026-01-21T12:07:58Z | user=sc2598 | action=VIEW-PAGE | page=/courses/cs439"
]

for t in tests:
    print(parse_fields(t))
    
        

```
